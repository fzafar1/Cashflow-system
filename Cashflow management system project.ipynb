```python
from tkinter import *


class Cash_Flow:
    def __init__(self,root):
        self.root = root
        self.root.geometry("1300x700+0+0")
        self.root.maxsize(width = 1280,height = 700)
        self.root.minsize(width = 1280,height = 700)
        self.root.title("Cash Flow System")
        
        #====================Variables========================#
        self.comp_name = StringVar()
        self.date = StringVar()
        self.begin = IntVar()
        self.netincome = IntVar()
        self.dep = IntVar()
        self.AP = IntVar()
        self.AR = IntVar()
        self.equipment= IntVar()
        self.dividend = IntVar()
        self.totalss=IntVar()
       
        

    
        #===================================
        bg_color = "steel blue"
        fg_color = "white"
        lbl_color = 'white'
        #Title of App
        title = Label(self.root,text = "Cash Flow Management",bd = 12,relief = GROOVE,fg = fg_color,bg = bg_color,font=("times new roman",30,"bold"),pady = 3).pack(fill = X)

        #==========Company Frame==========#
        F1 = LabelFrame(text = "Company Details",font = ("time new roman",12,"bold"),fg = "gold",bg = bg_color,relief = GROOVE,bd = 10)
        F1.place(x = 0,y = 80,relwidth = 1)

        #===============Company Name===========#
        cname_lbl = Label(F1,text="Company Name ",bg = bg_color,fg = fg_color,font=("times new roman",15,"bold")).grid(row = 0,column = 0,padx = 10,pady = 5)
        cname_en = Entry(F1,bd = 8,relief = GROOVE,textvariable = self.comp_name)
        cname_en.grid(row = 0,column = 1,ipady = 4,ipadx = 30,pady = 5)

        #=================date==============#
        date_lbl = Label(F1,text = "Date",bg = bg_color,fg = fg_color,font = ("times new roman",15,"bold")).grid(row = 0,column = 2,padx = 20)
        date_en = Entry(F1,bd = 8,relief = GROOVE,textvariable = self.date)
        date_en.grid(row = 0,column = 3,ipady = 4,ipadx = 30,pady = 5)

        #====================cash at beginning==================#
        cbegin_lbl = Label(F1,text = "Cash At beginning",bg = bg_color,fg = fg_color,font = ("times new roman",15,"bold"))
        cbegin_lbl.grid(row = 0,column = 4,padx = 20)
        cbegin_en = Entry(F1,bd = 8,relief = GROOVE,textvariable = self.begin)
        cbegin_en.grid(row = 0,column = 5,ipadx = 30,ipady = 4,pady = 5)
        

        #==================operations Frame=====================#
        F2 = LabelFrame(self.root,text = 'Cash FLow From operating Activity',bd = 10,relief = GROOVE,bg = bg_color,fg = "gold",font = ("times new roman",13,"bold"))
        F2.place(x = 5,y = 180,width = 325,height = 380)

        #===========Frame Content
        net_lbl = Label(F2,font = ("times new roman",15,"bold"),fg = lbl_color,bg = bg_color,text = "Net Income")
        net_lbl.grid(row = 0,column = 0,padx = 10,pady = 20)
        net_en = Entry(F2,bd = 8,relief = GROOVE,textvariable = self.netincome)
        net_en.grid(row = 0,column = 1,ipady = 5,ipadx = 5)

        #=================
        dep_lbl = Label(F2,font = ("times new roman",15,"bold"),fg = lbl_color,bg = bg_color,text = "Depreciation")
        dep_lbl.grid(row = 1,column = 0,padx = 10,pady = 20)
        dep_en = Entry(F2,bd = 8,relief = GROOVE,textvariable = self.dep)
        dep_en.grid(row = 1,column = 1,ipady = 5,ipadx = 5)

        #=================
        ap_lbl = Label(F2,font = ("times new roman",15,"bold"),fg = lbl_color,bg = bg_color,text = "Account Payable")
        ap_lbl.grid(row = 2,column = 0,padx = 10,pady = 20)
        ap_en = Entry(F2,bd = 8,relief = GROOVE,textvariable = self.AP)
        ap_en.grid(row = 2,column = 1,ipady = 5,ipadx = 5)

        #=================
        ar_lbl = Label(F2,font = ("times new roman",15,"bold"),fg = lbl_color,bg = bg_color,text = "Account Receivable")
        ar_lbl.grid(row = 3,column = 0,padx = 10,pady = 20)
        ar_en = Entry(F2,bd = 8,relief = GROOVE,textvariable = self.AR)
        ar_en.grid(row = 3,column = 1,ipady = 5,ipadx = 5)



        #==================investing Frame=====================#
        F2 = LabelFrame(self.root,text = 'Cash Flow From Investing Activities',bd = 10,relief = GROOVE,bg = bg_color,fg = "gold",font = ("times new roman",13,"bold"))
        F2.place(x = 330,y = 180,width = 325,height = 380)

        #===========Frame=======
        eq_lbl = Label(F2,font = ("times new roman",15,"bold"),fg = lbl_color,bg = bg_color,text = "Purchase Of Equipment")
        eq_lbl.grid(row = 0,column = 0,padx = 10,pady = 20)
        eq_en = Entry(F2,bd = 8,relief = GROOVE,textvariable = self.equipment)
        eq_en.grid(row = 0,column = 1,ipady = 5,ipadx = 5)

    

        #==================financing frame=====================#

        F2 = LabelFrame(self.root,text = "Financing Activities",bd = 10,relief = GROOVE,bg = bg_color,fg = "gold",font = ("times new roman",13,"bold"))
        F2.place(x = 655,y = 180,width = 325,height = 380)

        #===========Frame========
        dd_lbl = Label(F2,font = ("times new roman",15,"bold"),fg = lbl_color,bg = bg_color,text = "Dividends")
        dd_lbl.grid(row = 0,column = 0,padx = 10,pady = 20)
        dd_en = Entry(F2,bd = 8,relief = GROOVE,textvariable = self.dividend)
        dd_en.grid(row = 0,column = 1,ipady = 5,ipadx = 5)


        #===================Cash flow statement================
        F3 = Label(self.root,bd = 10,relief = GROOVE)
        F3.place(x = 960,y = 180,width = 325,height = 380)
        #===========
        c_title = Label(F3,text = "Cash Flow Statement",font = ("Lucida",13,"bold"),bd= 7,relief = GROOVE)
        c_title.pack(fill = X)

        #============
        scroll_y = Scrollbar(F3,orient = VERTICAL)
        self.txt = Text(F3,yscrollcommand = scroll_y.set)
        scroll_y.pack(side = RIGHT,fill = Y)
        scroll_y.config(command = self.txt.yview)
        self.txt.pack(fill = BOTH,expand = 1)

        #===========Buttons Frame=============
        F4 = LabelFrame(self.root,bd = 10,relief = GROOVE,bg = bg_color,fg = "gold",font = ("times new roman",13,"bold"))
        F4.place(x = 0,y = 560,relwidth = 1,height = 145)

        #===================
        netc_lbl = Label(F4,font = ("times new roman",15,"bold"),fg = lbl_color,bg = bg_color,text = "Net cash from operations")
        netc_lbl.grid(row = 0,column = 0,padx = 10,pady = 0)
        netc_en = Entry(F4,bd = 8,relief = GROOVE,textvariable = self.totalss)
        netc_en.grid(row = 0,column = 1,ipady = 2,ipadx = 5)


        #====================
        totals_btn = Button(F4,text = "Total",bg = bg_color,fg = fg_color,font=("lucida",12,"bold"),bd = 7,relief = GROOVE,command = self.total)
        totals_btn.grid(row = 1,column = 4,ipadx = 20,padx = 30)

        #========================
        gen_btn = Button(F4,text = "Generate",bg = bg_color,fg = fg_color,font=("lucida",12,"bold"),bd = 7,relief = GROOVE,command = self.generate)
        gen_btn.grid(row = 1,column = 5,ipadx = 20)

        #====================
        clear_btn = Button(F4,text = "Clear",bg = bg_color,fg = fg_color,font=("lucida",12,"bold"),bd = 7,relief = GROOVE,command = self.clear)
        clear_btn.grid(row = 1,column = 6,ipadx = 20,padx = 30)

        #======================
        exit_btn = Button(F4,text = "Exit",bg = bg_color,fg = fg_color,font=("lucida",12,"bold"),bd = 7,relief = GROOVE,command = self.exit)
        exit_btn.grid(row = 1,column = 7,ipadx = 20)
        


#Function to get total:
    def total(self):
        self.total_OA = (
            self.netincome.get()+
            self.AP.get()-
            self.AR.get()+
            self.dep.get()
        )
        self.totalss.set(str(self.total_OA))
           
        
class Cash_Statement(Cash_Flow):

    def welcome(self):
       
        self.txt.insert(END,"     ")
        self.txt.insert(END,f"\n {str(self.comp_name.get())}")
        self.txt.insert(END,f"\n {str(self.date.get())}")
        self.txt.insert(END,"\n==================================")
        self.txt.insert(END,f"\nCash at beginning of year: {str(self.begin.get())}")
        self.txt.insert(END,"\n___________________________________")
        
#Function to clear 
    def clear(self):
        self.txt.delete('1.0',END)


    def generate(self):
        self.welcome()
        if self.netincome.get() != 0:
            self.txt.insert(END,"\nCash Flow from Operating Activity")
            self.txt.insert(END,"\n______________")
            self.txt.insert(END,f"\nNet income              {self.netincome.get()}")
        if self.dep.get() != 0:
            self.txt.insert(END,f"\nDepreciation            {self.dep.get()}")
        if self.AP.get() != 0:
            self.txt.insert(END,f"\nAccount Payable         {self.AP.get()}")
        if self.AR.get() != 0:
            self.txt.insert(END,f"\nAccount Receivables     {self.AR.get()}")
            self.txt.insert(END,f"\nNet Cash                {self.totalss.get()}")
            self.txt.insert(END,"\n_________________________________")
        if self.equipment.get() != 0:
            self.txt.insert(END,"\nCash Flow From Investing Activity")
            self.txt.insert(END,"\n______________")
            self.txt.insert(END,f"\nPurchase of Equipment   {self.equipment.get()}")
            self.txt.insert(END,"\n_________________________________")
        if self.dividend.get() != 0:
            self.txt.insert(END,"\nCash Flow From Financing Activity")
            self.txt.insert(END,"\n______________")
            self.txt.insert(END,f"\nDividends       {self.dividend.get()}")
            self.txt.insert(END,"\n_________________________________")
        self.txt.insert(END,f"\nNet income          {(self.netincome.get()+self.dep.get()+self.AP.get()-self.AR.get())-self.equipment.get()-self.dividend.get()}")
        self.txt.insert(END,"\n______________________________")
        self.txt.insert(END,f"\n Cash at end of period : {((self.totalss.get())-self.equipment.get()-self.dividend.get())+self.begin.get()}")


    #Function to exit
    def exit(self):
        self.root.destroy()


root = Tk()
object = Cash_Statement(root)
root.mainloop()



```
